```python
import matplotlib.pyplot as plt
#sample data
temperature = [25,28,32,35,38,42,45,30,33,36,37] #temp in celsius
ice_cream_sales = [100,150,200,250,300,350,400,200,250,300,350]# in units

#scatter plot
plt.figure(figsize=(8,6))
plt.scatter(temperature,ice_cream_sales,color='#356f56',edgecolor='black',alpha=0.8)
plt.title("Temperature vs ice cream sales")
plt.xlabel("Temperature(Celsius)")
plt.ylabel("Ice cream sales(units)")
plt.grid(alpha=0.3)
plt.show()
```
